```python
import pandas as pd
path = 'columns_format.xlsx'

df = pd.read_excel(path, sheet_name='Sheet1')
df
```

```python
cols = list(df.columns)
cols = [col.split() for col in cols]
cols = [
    [x for x in col if (("200" not in x) and ("לעומת" not in x) and ("תשס" not in x))] 
    for col in cols
    ]

cols = [" ".join(col) for col in cols]
```

```python
df.columns = cols
```

```python
df.to_excel('columns_format.xlsx', index=False)
```
